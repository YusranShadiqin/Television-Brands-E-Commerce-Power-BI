```python
#Menampilkan semua unique values disetiap kolom
for column in data.columns:
    print(" ")
    print("Unique values from", column, 'is ', data[column].unique())
```

### Dikarenakan pada kolom "Operating System" dan "Rating" terdapat data NaN, maka data tersebut perlu diganti

```python
#Mengganti data NaN Operating System menjadi Unknown
data["Operating System"] = data["Operating System"].fillna("Unknown")
```

```python
#Melakukan validasi apakah penggantian data telah berhasil
data["Operating System"].unique()
```

```python
#Mengganti data NaN Rating menjadi 0
data['Rating'] = data['Rating'].fillna(0)
```

```python
#Melakukan validasi apakah penggantian data telah berhasil
data['Rating'].unique()
```

```python
#Melakukan validasi ulang apakah masih ada data NaN atau tidak
data.isnull().sum()
```

```python
#Melihat apakah ada data yang terduplikasi atau tidak
data.duplicated().sum()
```

```python
#Melihat semua data yang terduplikasi
data.loc[data.duplicated]
```

### Dikarenakan data yang terduplikasi tidak "Terduplikasi seperti biasanya", dan data adalah data unique maka tidak apa apa
